# **Constructor**

```python
class student:
  def __init__(self, name, age):
    self.name = name
    self.age = age

std = student('Ahmad', 15)
print(std.name, std.age)
```

```python
class student:
  def __init__(self, name, grade, percentage):
    self.name = name
    self.grade = grade
    self.percentage = percentage
    # The print statement was in a conflicting __init__ method.
    # If you intend to print student details, it's better to have a separate method for it.

std = student('Jawad', 11, 90)
print(std.__dict__)
```

Modify object property

```python
class student:
  def __init__(self, name, grade, percentage):
    self.name = name
    self.grade = grade
    self.percentage = percentage

std = student('Jawad', 11, 90)
print(std.__dict__)

std.percentage = 99
print(std.percentage)
```

Delete Object Property

```python
class student:
  def __init__(self, name, grade, percentage):
    self.name = name
    self.grade = grade
    self.percentage = percentage

std = student('Jawad', 11, 90)
print(std.__dict__)
del std.percentage
print(std.__dict__)
```

# **Features in OOP**

# Abstraction

```python
#Abstraction - Hiding un-necessary details from users through classes and methods.

class student:
  def __init__(self, name, grade, percentage):
    self.name = name
    self.grade = grade
    self.percentage = percentage

  def stu_det(self):
    print(f"{self.name} is in class {self.grade} with {self.percentage+2}% ")
#The above method is hidden from the user that's why it is called abstraction.

std = student('Ali', 11, 96)
std1 = student('Sara', 12, 88)

std.stu_det()
std1.stu_det()
```

# **Encapsulation**

```python
#Restrict access to certain attributes or methods to protect data and enforce controll access.
class student:
  def __init__(self, name, grade, percentage):
    self.name = name
    self.grade = grade
    self.__percentage = percentage #Double underscore limits access

  def get_percentage(self): #Method to access private attribute
    return self.__percentage

  def stu_det(self):
    print(f"{self.name} is in class {self.grade} with {self.percentage}% ")


std = student('Ali', 11, 96)
std1 = student('Sara', 12, 88)

print(std.get_percentage())
print(std1.get_percentage())
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age

  def get_age(self):
    return self.__age

p1 = Person('John', 19)

print(p1.name)
print(p1.get_age())
```
