```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
# Plot a scatter between total_bill and tip.
sns.relplot(x = 'total_bill', y = 'tip', data = data)
```

```python
# The hue parameter provided an extra piece of information e.g for smoker property there will be two different colors in dots that
# separate a smoker from a non smoker.

sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = data)
```

```python
''' The style parameter also provides extra information e.g it differentiate between gender like a cross dot for female and a circle
 dot for male.'''

sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style  = 'sex', data = data)

```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style  = 'sex', size = 'size', data = data)
```

```python
# Using scatterplot
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'smoker', style  = 'sex', data = data)
```

# **Linear Regression model predicting salaries**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "Experience": [1, 2, 3, 4, 5],
    "Salary": [30000, 35000, 40000, 45000, 50000]
}

df = pd.DataFrame(data)
```

```python
X = df[["Experience"]]   # input
y = df["Salary"]         # output
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[7]])

print("Predicted Salary:", prediction[0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title('Experience VS Salary')
plt.show()
```
